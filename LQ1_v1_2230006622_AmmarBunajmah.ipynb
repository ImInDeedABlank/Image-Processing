```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

### Task#1: Load/read images cat.jpeg and cameraman.tif [1 mark]

```python
# Load/read images cat.jpeg and cameraman.tif [1 mark]

cat = cv2.imread("cat.jpeg", cv2.IMREAD_GRAYSCALE)
cameraman = cv2.imread("cameraman.tif", cv2.IMREAD_GRAYSCALE)

```

### Task#2: Add the two images, and display the result [2 marks]

```python
# Add the two images, and display the result [2 marks]
cameraman = cv2.resize(cameraman, (cat.shape[1], cat.shape[0]))
result = cv2.add(cat, cameraman)
plt.imshow(result, cmap="gray")
plt.axis("off")
plt.show()
```

### Task#3: Using kernel1 and kernel2 apply convolution to cameraman.tif, and display the output image [2 marks]

```python
#Using kernel1 and kernel2 apply convolution to cameraman.tif, and display the output image [2 marks]
kernel1 = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]])

kernel2 = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

kernel1_result = cv2.filter2D(cameraman, -1, kernel1)
kernel2_result = cv2.filter2D(cameraman, -1, kernel2)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(kernel1_result, cmap="gray")
plt.title("Kernel 1 Result")
plt.axis("off")

# This is horziontal edge detection?
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(kernel2_result, cmap="gray")
plt.title("Kernel 2 Result")
plt.axis("off")

# This is vertical edge detection !! nice wow
```

```python
#if we add both we get full edge detection result

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(kernel1_result + kernel2_result, cmap="gray")
```
